```python
import pandas as pd
import datetime
import numpy as np
import os
```

```python
#get csv files 202009, 202008, 202011,202010 (last four months)
filepath_aug = os.path.join("Resources","202008-citibike-tripdata.csv")
filepath_sep = os.path.join("Resources","202009-citibike-tripdata.csv")
filepath_oct = os.path.join("Resources","202010-citibike-tripdata.csv")
filepath_nov = os.path.join("Resources","202011-citibike-tripdata.csv")

nycbike_2020aug = pd.read_csv(filepath_aug)
nycbike_2020sep = pd.read_csv(filepath_sep)
nycbike_2020oct = pd.read_csv(filepath_oct)
nycbike_2020nov = pd.read_csv(filepath_nov)
```

```python
#check files 
nycbike_2020aug.head(10)
nycbike_2020sep.head(10)
nycbike_2020oct.head(10) 
nycbike_2020nov.head(10) 
```

```python
#concatenate csv files
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html
nycbike_aug_sep_oct_nov = pd.concat([nycbike_2020aug,nycbike_2020sep,nycbike_2020oct,nycbike_2020nov])
```

```python
#output concatenation to csv file
filepath_2020_aug_sep_oct_nov = os.path.join("Resources","2020-citibike-tripdata.csv")
nycbike_aug_sep_oct_nov.to_csv(filepath_2020_aug_sep_oct_nov, index=False)
nycbike_aug_sep_oct_nov.tail(10)
```
